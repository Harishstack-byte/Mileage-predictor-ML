```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X.replace('?', np.nan, inplace=True)
X = X.apply(pd.to_numeric, errors='coerce')

X.fillna(X.mean(), inplace=True)

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train the model
lr = LinearRegression()
lr.fit(X_train, Y_train)


```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy  as np

Y_predict = lr.predict(X_test)
lr_r2 = r2_score(Y_test, Y_predict) * 100
lr_mse = mean_squared_error(Y_test, Y_predict)
lr_rmse = np.sqrt(lr_mse)
lr_mae = mean_absolute_error(Y_test, Y_predict)

print("Linear Regression Model Evaluation:")
print("The R² Score:",lr_r2)
print("Mean Squared Error (MSE):",lr_mse)
print("Root Mean Squared Error (RMSE):",lr_rmse)
print("Mean Absolute Error (MAE):",lr_mae)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
data = pd.read_csv("MPG_dataset.csv")
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
data.dropna(inplace=True)
features = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year']
X = data[features]
Y = data['mpg']
# Scale features for Linear Regression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train models
lr = LinearRegression()
lr.fit(X_scaled, Y)

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, Y)

#input from user for prediction
print("🔧 Enter car details to predict mileage:")
cylinders = float(input("Cylinders: "))
displacement = float(input("Displacement: "))
horsepower = float(input("Horsepower: "))
weight = float(input("Weight: "))
acceleration = float(input("Acceleration: "))
model_year = float(input("Model Year: "))
user_input = pd.DataFrame([[cylinders, displacement, horsepower, weight, acceleration, model_year]],
                          columns=features)

# Predict Linear Regression 
user_input_scaled = scaler.transform(user_input)
lr_prediction = lr.predict(user_input_scaled)

# Predict Random Forest 
rf_prediction = rf.predict(user_input)

print("Mileage Predictions:")
print("Linear Regression:",lr_prediction,"MPG")
print("Random Forest:", rf_prediction,"MPG")
```
